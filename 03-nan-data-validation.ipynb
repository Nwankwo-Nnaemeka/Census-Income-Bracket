```python
config_box = read_yaml("config.yaml")
schema_box = read_yaml("schema.yaml")
config = config_box.data_validation
schema = schema_box.FEATURES
```

```python

class DataValidation:
    def __init__(self, config, schema):
        self.config = config
        self.schema = schema

    def validate_columns(self):
        validation_status = None
        
        data = pd.read_csv(self.config.train_data_dir)
        all_columns = list(data.columns)
        schema = self.schema.keys()
        schema_values = self.schema.values()
        for column in all_columns:
            if column not in schema:
                validation_status = False
                with open(self.config.STATUS_FILE, 'w') as file:
                    file.write(f"validation status: {validation_status}")
                
            else:
                validation_status = True
                with open(self.config.STATUS_FILE, 'w') as file:
                    file.write(f"validation status: {validation_status}")
        return validation_status  
```

```python
valid_data = DataValidation(config, schema)
```
